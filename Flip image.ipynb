```python
import os
import cv2
import shutil
```

```python
directory_path = './mp_four10k'
output_directory ='./flp_four'
```

```python
def flip_images_in_directory(directory_path, output_directory, flip_horizontal=True, flip_vertical=False):
    os.makedirs(output_directory, exist_ok=True)
    
    image_files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]
    
    for img_file in image_files:
        img_path = os.path.join(directory_path, img_file)
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Unable to read image '{img_path}'")
            continue
        
        flipped_img = None
        if flip_horizontal:
            flipped_img = cv2.flip(img, 1)
        elif flip_vertical:
            flipped_img = cv2.flip(img, 0)
        else:
            print("Error: No flipping direction specified.")
            continue
        
        flipped_img_filename = "flp_" + img_file
        
        flipped_img_path = os.path.join(output_directory, flipped_img_filename)
        cv2.imwrite(flipped_img_path, flipped_img)
        
        print(f"Flipped image saved: {flipped_img_path}")
```

```python
flip_images_in_directory(directory_path, output_directory, flip_horizontal=True)
```
